```python
# Augmentation Microscope

class Microscope(ImageOnlyTransform):
    def __init__(self, p: float = 0.5, always_apply=False):
        super().__init__(always_apply, p)

    def apply(self, img, **params):
        if random.random() < self.p:
            # Ensure img is in uint8 format
            if img.dtype != np.uint8:
                img = (img * 255).astype(np.uint8)

            # Create a circular mask with the same dimensions as the image
            mask = np.ones_like(img, dtype=np.uint8) * 255
            center = (img.shape[1] // 2, img.shape[0] // 2)
            radius = random.randint(img.shape[0] // 2 - 3, img.shape[0] // 2 + 15)
            cv2.circle(mask, center, radius, (0, 0, 0), -1)

            # Invert mask for correct multiplication
            mask_inv = 1 - (mask // 255)

            # Apply mask to the image
            img = cv2.bitwise_and(img, img, mask=mask_inv[:, :, 0])

        return img
```

```python
# Augmentation Hair Noise

def get_bezier_curve(points, n_points = 100):
    points = np.array(points)
    n = len(points)
    binomial_coeffs = np.array([np.math.comb(n-1, i) for i in range(n)])

    def bernstein_polynomial(i, t, n):
        return binomial_coeffs[i] * (t ** i) * ((1 - t) ** (n - i))

    curve = np.zeros((n_points, 2))
    t_values = np.linspace(0, 1, n_points)

    for i in range(n):
        curve += np.outer(bernstein_polynomial(i, t_values, n-1), points[i])

    return curve

class AddHairNoise(ImageOnlyTransform):
    def __init__(self, hair_min = 5, hair_max = 20, hair_length = 30, hair_thickness = 1, n_points = 100, always_apply = False, p = 0.5):
        super(AddHairNoise, self).__init__(always_apply, p)
        self.hair_min = hair_min
        self.hair_max = hair_max
        self.hair_length = hair_length
        self.hair_thickness = hair_thickness
        self.n_points = n_points

    def apply(self, image, **params):

        if random.random() > self.p:
            return image

        image_with_hair = image.copy()
        height, width, _ = image.shape

        hair_count = random.randint(self.hair_min, self.hair_max)

        for _ in range(hair_count):
            x_start = np.random.randint(0, width)
            y_start = np.random.randint(0, height)

            control_points = [(x_start, y_start)]
            for _ in range(2):
                x_control = x_start + np.random.randint(-self.hair_length, self.hair_length)
                y_control = y_start + np.random.randint(-self.hair_length, self.hair_length)
                control_points.append((x_control, y_control))
            x_end = x_start + np.random.randint(-self.hair_length, self.hair_length)
            y_end = y_start + np.random.randint(-self.hair_length, self.hair_length)
            control_points.append((x_end, y_end))

            curve_points = get_bezier_curve(control_points, self.n_points)

            for i in range(len(curve_points) - 1):
                cv2.line(image_with_hair,
                            (int(curve_points[i][0]), int(curve_points[i][1])),
                            (int(curve_points[i+1][0]), int(curve_points[i+1][1])),
                            (0, 0, 0), self.hair_thickness
                        )

        return image_with_hair
```

```python
class MelanomaDataset(Dataset):
    def __init__(self, df, meta_features, hdf, transform = None, hsv_transform = None, mode = 'test'):
        self.isic_ids = df['isic_id'].values
        self.df = df.drop('isic_id', axis = 1).reset_index(drop = True)
        self.use_meta = meta_features is not None
        self.meta_features = meta_features
        self.hdf5_file = h5py.File(hdf, mode = 'r')
        self.transform = transform
        self.hsv_transform = hsv_transform
        self.mode = mode

    def __len__(self):
        return len(self.isic_ids)

    def hair_remove(self, image):
        # convert grayScale
        grayScale = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))

        # remove black hair
        blackhat = cv2.morphologyEx(grayScale, cv2.MORPH_BLACKHAT, kernel)
        _, blackhat_thresh = cv2.threshold(blackhat, 10, 255, cv2.THRESH_BINARY)

        # remove white hair
        invert_gray = cv2.bitwise_not(grayScale)
        whitehat = cv2.morphologyEx(invert_gray, cv2.MORPH_BLACKHAT, kernel)
        _, whitehat_thresh = cv2.threshold(whitehat, 10, 255, cv2.THRESH_BINARY)

        combined_thresh = cv2.bitwise_or(blackhat_thresh, whitehat_thresh)

        result = cv2.inpaint(image, combined_thresh, 1, cv2.INPAINT_TELEA)
        return result

    def remove_markers(self, image):
        hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

        lower_purple = np.array([125, 50, 50])
        upper_purple = np.array([155, 255, 255])

        mask = cv2.inRange(hsv, lower_purple, upper_purple)

        mask = cv2.GaussianBlur(mask, (5, 5), 0)

        _, mask = cv2.threshold(mask, 1, 255, cv2.THRESH_BINARY)

        result = cv2.inpaint(image, mask, 3, cv2.INPAINT_TELEA)

        return result

    def remove_black_borders(self, image, threshold_value=10):
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

        _, thresh = cv2.threshold(gray, threshold_value, 255, cv2.THRESH_BINARY)

        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        if contours:
            x, y, w, h = cv2.boundingRect(contours[0])
            for contour in contours:
                x_, y_, w_, h_ = cv2.boundingRect(contour)
                x = min(x, x_)
                y = min(y, y_)
                w = max(w, w_ + x_ - x)
                h = max(h, h_ + y_ - y)

            cropped_image = image[y:y+h, x:x+w]
            return cropped_image
        else:
            return image

    def get_labels(self):
        return self.df['target'].values

    def __getitem__(self, idx):
        isic_id = self.isic_ids[idx]
        # image = np.array(self.hdf5_file[isic_id])
        # image = np.array(Image.open(io.BytesIO(image)))
        image = Image.open(BytesIO(self.hdf5_file[isic_id][()]))
        image = np.array(image)

        image = self.hair_remove(image)
        image = self.remove_markers(image)
        image = self.remove_black_borders(image)

#         # enhance image
#         image = cv2.medianBlur(image, 5)

#         z = image.reshape((-1, 3))
#         z = np.float32(z)

#         criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
#         k = 8
#         ret, label, center = cv2.kmeans(z, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

#         center = np.uint8(center)
#         res = center[label.flatten()]
#         image = res.reshape((image.shape))

#         hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV) # COLOR_RGB2HSV, COLOR_BGR2HSV
#         h, s, v = cv2.split(hsv)
#         clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
#         h_clahe = clahe.apply(h)
#         s_clahe = clahe.apply(s)
#         v_clahe = clahe.apply(v)
#         enhanced_hsv = cv2.merge((h_clahe, s_clahe, v_clahe))
#         image = cv2.cvtColor(enhanced_hsv, cv2.COLOR_LAB2RGB) # COLOR_LAB2BGR # COLOR_HSV2RGB

        hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV) # 6 channels # COLOR_RGB2HSV, COLOR_BGR2HSV

        if self.transform is not None:
            res = self.transform(image = image)['image']
            image = torch.tensor(res, dtype = torch.float32)
        else:
            image = torch.tensor(image, dtype = torch.float32)

        if self.hsv_transform is not None:
            hsv_image = self.hsv_transform(image = hsv_image)['image']

            # Concatenate channel dimension -> (6, H, W)
            image = torch.cat((image, hsv_image), dim = 0)

        # image = image.permute(2, 0, 1)

        if self.use_meta:
            data = (torch.tensor(image, dtype = torch.float32), torch.tensor(self.df[self.meta_features].iloc[idx], dtype = torch.float32))
        else:
            data = torch.tensor(image, dtype = torch.float32)

        if self.mode == 'test':
            return data
        else:
            return data, self.df.iloc[idx].target # torch.tensor(self.df.iloc[idx].target, dtype = torch.long)
```

```python
def data_processor(train_df=None, test_df=None, numerical_variables=None, categorical_variables=None, is_scaled=False, is_encoded=False, scaler_path = None, encoder_path=None):
    meta_variables = []

    if is_scaled:
        if train_df is not None:
            scaler = StandardScaler() # StandardScaler() # MinMaxScaler()

            scaler = scaler.fit(train_df[numerical_variables])
            train_df[numerical_variables] = scaler.transform(train_df[numerical_variables])
            joblib.dump(scaler, 'scaler.pkl')

            train_df[numerical_variables] = train_df[numerical_variables].replace([np.inf, -np.inf], np.nan)
            train_df[numerical_variables] = train_df[numerical_variables].fillna(train_df[numerical_variables].median())

        if test_df is not None:

            scaler = joblib.load(scaler_path)
            test_df[numerical_variables] = scaler.transform(test_df[numerical_variables])

            test_df[numerical_variables] = test_df[numerical_variables].replace([np.inf, -np.inf], np.nan)
            test_df[numerical_variables] = test_df[numerical_variables].fillna(test_df[numerical_variables].median())

    if is_encoded:
        if train_df is not None:
            encoder = OneHotEncoder(sparse_output=False, dtype=np.int32, handle_unknown='ignore')
            encoder.fit(train_df[categorical_variables])

            joblib.dump(encoder, encoder_path)

            categorical_variables_updated = [f'onehot_{i}' for i in range(len(encoder.get_feature_names_out()))]

            train_df[categorical_variables_updated] = encoder.transform(train_df[categorical_variables])
            train_df[categorical_variables_updated] = train_df[categorical_variables_updated].astype('uint8')

            train_df = train_df.drop(categorical_variables, axis=1)
            meta_variables = numerical_variables + categorical_variables_updated

        if test_df is not None:
            encoder = joblib.load(encoder_path)

            categorical_variables_updated = [f'onehot_{i}' for i in range(len(encoder.get_feature_names_out()))]

            test_df[categorical_variables_updated] = encoder.transform(test_df[categorical_variables])
            test_df[categorical_variables_updated] = test_df[categorical_variables_updated].astype('uint8')

            test_df = test_df.drop(categorical_variables, axis=1)
            meta_variables = numerical_variables + categorical_variables_updated

    else:
        # les = {}
        # if train_df is not None:
        #     for cat_col in categorical_variables:
        #         le = LabelEncoder()
        #         le.fit(train_df[cat_col].astype(str))
        #         train_df[cat_col] = le.transform(train_df[cat_col].astype(str))
        #         train_df[cat_col] = train_df[cat_col].astype('category')
        #         les[cat_col] = le

        #     joblib.dump(les, encoder_path)  # Save the label encoders for future use
        #     meta_variables = numerical_variables + categorical_variables

        # if test_df is not None:
        #     les = joblib.load(encoder_path)  # Load the saved label encoders
        #     for cat_col in categorical_variables:
        #         le = les[cat_col]
        #         test_df[cat_col] = test_df[cat_col].astype(str)
        #         test_df[cat_col] = test_df[cat_col].apply(lambda x: x if x in le.classes_ else 'unknown')
        #         test_df[cat_col] = test_df[cat_col].map(lambda x: le.transform([x])[0] if x in le.classes_ else -1)
        #         test_df[cat_col] = test_df[cat_col].astype('category')
        #     meta_variables = numerical_variables + categorical_variables

        if train_df is not None:

          encoder = OrdinalEncoder(
            categories='auto',
            dtype=int,
            handle_unknown='use_encoded_value',
            unknown_value=-2,
            encoded_missing_value=-1,
          )

          train_df[categorical_variables] = train_df[categorical_variables].astype(str)

          encoder.fit(train_df[categorical_variables])
          train_df[categorical_variables] = encoder.transform(train_df[categorical_variables])

          joblib.dump(encoder, 'encoder.pkl')

        if test_df is not None:
          test_df[categorical_variables] = test_df[categorical_variables].astype(str)

          encoder = joblib.load(encoder_path)
          test_df[categorical_variables] = encoder.transform(test_df[categorical_variables])

        meta_variables = numerical_variables + categorical_variables

    return train_df, test_df, meta_variables
```

```python
def get_dataloader(dataset, split = 'test'):

    if split == 'train':
        labels = dataset.get_labels()
        class_weights = torch.tensor(compute_class_weight(class_weight = 'balanced', classes = np.unique(labels), y = labels))
        samples_weights = class_weights[labels]
        sampler = WeightedRandomSampler(weights = samples_weights, num_samples = len(samples_weights), replacement = True)
        dataloader = DataLoader(dataset, batch_size = 128, sampler = sampler, shuffle = False, num_workers=4, pin_memory=True) # sampler = sampler, prefetch_factor=2, persistent_workers=True

    elif split == 'val' or split == 'test':
        dataloader = DataLoader(dataset, batch_size = 128, shuffle = False, num_workers=4, pin_memory=True) # sampler = SequentialSampler(dataset), prefetch_factor=2, persistent_workers=True

    return dataloader
```

```python
def get_fold(fold_idx, train_indices, val_indices, df, hdf_train_path, meta_features = None, binary_output = False):

    df_train, df_valid = df.iloc[train_indices], df.iloc[val_indices]

    train_dataset = MelanomaDataset(df = df_train, meta_features = meta_features, hdf = hdf_train_path, transform = get_rgb_transform(224, 224)[0], mode = 'train') # hsv_transform = get_hsv_transform(224, 224)[0],
    val_dataset = MelanomaDataset(df = df_valid, meta_features = meta_features, hdf = hdf_train_path, transform = get_rgb_transform(224, 224)[1], mode = 'val') # hsv_transform = get_hsv_transform(224, 224)[1],

    train_dataloader = get_dataloader(train_dataset, split = 'train')
    val_dataloader = get_dataloader(val_dataset, split = 'val')

    num_epochs = 1 # 5, 35
    val_scores = []
    early_stopping_counter = 0
    best_score = 0.0
    n_meta_features = 0
    use_meta = False
    warmup = True

    if meta_features is not None:
        n_meta_features = len(meta_features)
        use_meta = True

    model = EffNetB3_Melanoma(out_dim = 1, n_meta_features = n_meta_features)
    criterion = nn.BCEWithLogitsLoss() # nn.BCEWithLogitsLoss() # nn.CrossEntropyLoss() # nn.BCELoss() # FocalLoss()
    optimizer = optim.AdamW(model.parameters(), lr = 1e-4)

    if warmup:

      scheduler_cosine = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=500, eta_min=1e-6)
      scheduler = GradualWarmupSchedulerV2(optimizer, multiplier=10, total_epoch=1, after_scheduler=scheduler_cosine)
    else:
      scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=500, eta_min=1e-6)

    device = CONFIG['device']

    for epoch in tqdm(range(1, num_epochs + 1), desc = 'Epochs'):
        print(time.ctime(), f'Fold {fold_idx}, Epoch {epoch}')

        train_loss, train_score, model = train_epoch(model, train_dataloader, optimizer, criterion, scheduler, device, use_meta, 1, epoch, binary_output) # 1, 4

        val_loss, val_score = val_epoch(model, val_dataloader, criterion, p_auc_tpr, device, use_meta, binary_output)

        tqdm.write(f'Epoch {epoch}/{num_epochs}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Val score: {val_score:.4f}')
        val_scores.append(val_score)

        if val_score > best_score:
            best_score = val_score
            early_stopping_counter = 0
            torch.save(model.state_dict(), f'best_fold_{fold_idx + 1}.pth')
        else:
            early_stopping_counter += 1

        if early_stopping_counter >= 5:
            print('Early stopping triggered.')
            break

        if val_score > best_score:
            best_score = val_score
            torch.save(model.state_dict(), f'best_fold_{fold_idx + 1}.pth')
```

```python
sigmoid = nn.Sigmoid()

class Swish(torch.autograd.Function):
    @staticmethod
    def forward(ctx, i):
        result = i * sigmoid(i)
        ctx.save_for_backward(i)
        return result
    @staticmethod
    def backward(ctx, grad_output):
        i = ctx.saved_variables[0]
        sigmoid_i = sigmoid(i)
        return grad_output * (sigmoid_i * (1 + i * (1 - sigmoid_i)))


class Swish_Module(nn.Module):
    def forward(self, x):
        return Swish.apply(x)

class GeM(nn.Module):
    def __init__(self, p=3, eps=1e-6):
        super(GeM, self).__init__()
        self.p = nn.Parameter(torch.ones(1)*p)
        self.eps = eps

    def forward(self, x):
        return self.gem(x, p=self.p, eps=self.eps)

    def gem(self, x, p=3, eps=1e-6):
        return F.avg_pool2d(x.clamp(min=eps).pow(p), (x.size(-2), x.size(-1))).pow(1./p)

    def __repr__(self):
        return self.__class__.__name__ + \
                '(' + 'p=' + '{:.4f}'.format(self.p.data.tolist()[0]) + \
                ', ' + 'eps=' + str(self.eps) + ')'

class EffNetB3_Melanoma(nn.Module):
    def __init__(self, out_dim = 1, n_meta_features = 0, n_meta_dim = [512, 128], pretrained = False):
        super(EffNetB3_Melanoma, self).__init__()

        self.n_meta_features = n_meta_features

        self.model = timm.create_model('efficientnet_b3', pretrained = pretrained)

        self.dropouts = nn.ModuleList([
            nn.Dropout(0.5) for _ in range(5)
        ])
        in_ch = self.model.num_features

        if n_meta_features > 0:
            self.meta = nn.Sequential(
                nn.Linear(n_meta_features, n_meta_dim[0]),
                nn.BatchNorm1d(n_meta_dim[0]),
                nn.SiLU(),
                nn.Dropout(p = 0.3),
                nn.Linear(n_meta_dim[0], n_meta_dim[1]),
                nn.BatchNorm1d(n_meta_dim[1]),
                nn.SiLU(),
            )

            in_ch += n_meta_dim[1]

        self.fc = nn.Linear(in_ch, out_dim)
        self.model.classifier = nn.Identity()

    def extract(self, x):
        x = self.model(x)
        return x

    def forward(self, x, x_meta = None):
        x = self.extract(x).squeeze(-1).squeeze(-1)

        if self.n_meta_features > 0 and x_meta is not None:
            x_meta = self.meta(x_meta)
            x = torch.cat((x, x_meta), dim = 1)

        # generalization through averages
        for i, dropout in enumerate(self.dropouts):
            if i == 0:
                out = self.fc(dropout(x))
            else:
                out += self.fc(dropout(x))

        out /= len(self.dropouts)
        return out
```

```python
class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2, reduce=True):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduce = reduce

    def forward(self, inputs, targets):

        ce_loss = nn.CrossEntropyLoss()(inputs, targets)

        pt = torch.exp(-ce_loss)  # CE(pt) = -log(pt) --> -ce_loss = log(pt) --> exp(log(pt)) --> pt
        F_loss = self.alpha * (1-pt)**self.gamma * ce_loss

        if self.reduce:
            return torch.mean(F_loss)
        else:
            return F_loss
```

```python
def comp_score_v2(targets, predictions, min_tpr=0.80):
    # Ensure the inputs are numpy arrays for processing
    targets = np.array(targets)
    predictions = np.array(predictions)

    v_gt = abs(targets - 1)  # Assuming 'targets' are 0s and 1s
    v_pred = 1.0 - predictions  # Inverting predictions if necessary

    max_fpr = abs(1 - min_tpr)
    partial_auc_scaled = roc_auc_score(v_gt, v_pred, max_fpr=max_fpr)

    # Adjust scale from [0.5, 1.0] to [0.5 * max_fpr**2, max_fpr]
    partial_auc = 0.5 * max_fpr**2 + (max_fpr - 0.5 * max_fpr**2) / (1.0 - 0.5) * (partial_auc_scaled - 0.5)
    return partial_auc

from sklearn.metrics import roc_curve, auc, roc_auc_score

def score_array(solution: np.array, submission: np.array, min_tpr: float = 0.80) -> float:
    # Check submission is numeric
    if not np.issubdtype(submission.dtype, np.number):
        raise ValueError('Submission target column must be numeric')

    # Rescale the target: set 0s to 1s and 1s to 0s (since sklearn only has max_fpr)
    v_gt = np.abs(solution - 1)

    # Flip the submissions to their complements
    v_pred = -1.0 * submission

    max_fpr = abs(1 - min_tpr)

    # Using sklearn.metric functions: (1) roc_curve and (2) auc
    fpr, tpr, _ = roc_curve(v_gt, v_pred, sample_weight=None)
    if max_fpr is None or max_fpr == 1:
        return auc(fpr, tpr)
    if max_fpr <= 0 or max_fpr > 1:
        raise ValueError("Expected min_tpr in range [0, 1), got: %r" % min_tpr)

def score(solution: pd.DataFrame, submission: pd.DataFrame, row_id_column_name: str, min_tpr: float=0.80) -> float:
    '''
    2024 ISIC Challenge metric: pAUC

    Given a solution file and submission file, this function returns the
    the partial area under the receiver operating characteristic (pAUC)
    above a given true positive rate (TPR) = 0.80.
    https://en.wikipedia.org/wiki/Partial_Area_Under_the_ROC_Curve.

    (c) 2024 Nicholas R Kurtansky, MSKCC

    Args:
        solution: ground truth pd.DataFrame of 1s and 0s
        submission: solution dataframe of predictions of scores ranging [0, 1]

    Returns:
        Float value range [0, max_fpr]
    '''

    del solution[row_id_column_name]
    del submission[row_id_column_name]

    # check submission is numeric
    if not pandas.api.types.is_numeric_dtype(submission.values):
        raise ParticipantVisibleError('Submission target column must be numeric')

    # rescale the target. set 0s to 1s and 1s to 0s (since sklearn only has max_fpr)
    v_gt = abs(np.asarray(solution.values)-1)

    # flip the submissions to their compliments
    v_pred = -1.0*np.asarray(submission.values)

    max_fpr = abs(1-min_tpr)

    # using sklearn.metric functions: (1) roc_curve and (2) auc
    fpr, tpr, _ = roc_curve(v_gt, v_pred, sample_weight=None)
    if max_fpr is None or max_fpr == 1:
        return auc(fpr, tpr)
    if max_fpr <= 0 or max_fpr > 1:
        raise ValueError("Expected min_tpr in range [0, 1), got: %r" % min_tpr)

    # Add a single point at max_fpr by linear interpolation
    stop = np.searchsorted(fpr, max_fpr, "right")
    x_interp = [fpr[stop - 1], fpr[stop]]
    y_interp = [tpr[stop - 1], tpr[stop]]
    tpr = np.append(tpr[:stop], np.interp(max_fpr, x_interp, y_interp))
    fpr = np.append(fpr[:stop], max_fpr)
    partial_auc = auc(fpr, tpr)

#     # Equivalent code that uses sklearn's roc_auc_score
#     v_gt = abs(np.asarray(solution.values)-1)
#     v_pred = np.array([1.0 - x for x in submission.values])
#     max_fpr = abs(1-min_tpr)
#     partial_auc_scaled = roc_auc_score(v_gt, v_pred, max_fpr=max_fpr)
#     # change scale from [0.5, 1.0] to [0.5 * max_fpr**2, max_fpr]
#     # https://math.stackexchange.com/questions/914823/shift-numbers-into-a-different-range
#     partial_auc = 0.5 * max_fpr**2 + (max_fpr - 0.5 * max_fpr**2) / (1.0 - 0.5) * (partial_auc_scaled - 0.5)

    return(partial_auc)

def p_auc_tpr(v_gt, v_pred, min_tpr=0.8, sample_weight=None):
    """Computes the area under the AUC above a minumum TPR.

    Args:
        v_gt: ground truth vector (1s and 0s)
        v_p: predictions vector of scores ranging [0, 1]
        min_tpr: minimum true positive threshold (sensitivity)

    Returns:
        Float value range [0, 1]
    """
    # if len(np.unique(v_gt)) != 2:
    #     raise ValueError(
    #         "Only one class present in y_true. ROC AUC score "
    #         "is not defined in that case."
    #     )

    # redefine the target. set 0s to 1s and 1s to 0s
    v_gt = abs(np.asarray(v_gt)-1)
    v_pred = abs(np.asarray(v_pred)-1)
    max_fpr = abs(1-min_tpr)

    # using sklearn.metric functions: (1) roc_curve and (2) auc
    fpr, tpr, _ = roc_curve(v_gt, v_pred, sample_weight=sample_weight)
    if max_fpr is None or max_fpr == 1:
        return auc(fpr, tpr)
    if max_fpr <= 0 or max_fpr > 1:
        raise ValueError("Expected min_tpr in range [0, 1), got: %r" % min_tpr)

    # Add a single point at max_fpr by linear interpolation
    stop = np.searchsorted(fpr, max_fpr, "right")
    x_interp = [fpr[stop - 1], fpr[stop]]
    y_interp = [tpr[stop - 1], tpr[stop]]
    tpr = np.append(tpr[:stop], np.interp(max_fpr, x_interp, y_interp))
    fpr = np.append(fpr[:stop], max_fpr)
    partial_auc = auc(fpr, tpr)
    return(partial_auc)
```

```python
def train_epoch(model, dataloader, optimizer, criterion, scheduler, device, use_meta, n_accumulate, epoch):
    # Initialize model for multi-GPU (if needed)
    # if torch.cuda.device_count() > 1:
    #     model = nn.DataParallel(model.to(device))
    # else:
    #     model = model.to(device)

    model = model.to(device)

    model.train()

    dataset_size = 0
    running_loss = 0.0
    running_auroc = 0.0
    train_loss = []

    scaler = torch.cuda.amp.GradScaler()

    bar = tqdm(enumerate(dataloader), desc='Training', total=len(dataloader), leave=False)

    for step, batch in bar:
        optimizer.zero_grad()

        if use_meta:
            (image_data, meta_data), target = batch
            image_data, meta_data, target = (
                image_data.to(device, dtype=torch.float),
                meta_data.to(device, dtype=torch.float),
                target.to(device, dtype=torch.float),
            )

            with torch.cuda.amp.autocast():
                output = model(image_data, meta_data).squeeze()

        else:
            image_data, target = batch
            image_data, target = (
                image_data.to(device, dtype=torch.float),
                target.to(device, dtype=torch.float),
            )

            with torch.cuda.amp.autocast():
                output = model(image_data).squeeze()

        loss = criterion(output, target)
        loss /= n_accumulate
        batch_size = target.size(0)
        target = target.type_as(output)
        train_loss.append(loss.item())
        # smooth_loss = sum(train_loss[-100:]) / min(len(train_loss), 100)

        scaler.scale(loss).backward()

        if (step + 1) % n_accumulate == 0:
            scaler.step(optimizer)
            scaler.update()

            if scheduler is not None:
                scheduler.step(epoch - 1)

        auroc = p_auc_tpr(target.detach().cpu().numpy(), output.detach().cpu().numpy())

        running_loss += loss.item() * batch_size * n_accumulate
        running_auroc += auroc * batch_size

        dataset_size += batch_size

        epoch_loss = running_loss / dataset_size
        epoch_auroc = running_auroc / dataset_size

        bar.set_postfix(Train_loss=epoch_loss, Train_auroc=epoch_auroc) # Smooth_loss=smooth_loss

    return epoch_loss, epoch_auroc, model
```

```python
def val_epoch(model, dataloader, criterion, scorer, device, use_meta, binary_output=False):
    model.eval()
    outputs = []
    targets = []
    dataset_size = 0
    running_loss = 0.0

    bar = tqdm(dataloader, desc='Validating', total=len(dataloader), leave=False)

    with torch.no_grad():
        for batch in bar:
            if use_meta:
                (image_data, meta_data), target = batch
                image_data, meta_data, target = (
                    image_data.to(device, dtype=torch.float),
                    meta_data.to(device, dtype=torch.float),
                    target.to(device, dtype=torch.float),
                )

                with torch.cuda.amp.autocast():
                    output = model(image_data, meta_data).squeeze()
                    output = output.sigmoid() if binary_output else output.softmax(1)
            else:
                image_data, target = batch
                image_data, target = (
                    image_data.to(device, dtype=torch.float),
                    target.to(device, dtype=torch.float),
                )

                with torch.cuda.amp.autocast():
                    output = model(image_data).squeeze()
                    output = output.sigmoid() if binary_output else output.softmax(1)

            loss = criterion(output, target)
            batch_size = target.size(0)

            target = target.type_as(output)
            outputs.append(output.cpu())
            targets.append(target.cpu())

            running_loss += loss.item() * batch_size
            dataset_size += batch_size

            epoch_loss = running_loss / dataset_size
            bar.set_postfix(Val_loss=epoch_loss)

    outputs = torch.cat(outputs)
    targets = torch.cat(targets)

    print('outputs:\n', outputs)
    print('\n')
    print('targets:\n', targets)

    try:
        val_scores = p_auc_tpr(targets.numpy(), outputs.numpy())
        print(f'p_auc_tpr: {val_scores}')
    except Exception as e:
        print(e)

    try:
        val_scores = scorer(outputs.numpy(), targets.numpy()).item()
        print(f'scorer: {val_scores}')
    except Exception as e:
        print(e)

    return epoch_loss, val_scores
```

```python
def load_model(model_class, checkpoint_path, device):
    model = model_class
    model.load_state_dict(torch.load(checkpoint_path, map_location = CONFIG['device']))
    return model

def inference(model, dataloader, device, use_meta=False, binary_output=True):
    predictions = []

    model.eval()

    with torch.no_grad():
        for batch in tqdm(dataloader, desc='Inference', leave=False):
            if use_meta:
                image_data, meta_data = batch
                image_data, meta_data = image_data.to(device, dtype=torch.float), meta_data.to(device, dtype=torch.float)
                output = model(image_data, meta_data).squeeze()
            else:
                image_data = batch.to(device, dtype=torch.float)
                output = model(image_data).squeeze()

            if binary_output:
                output = output.sigmoid()
            else:
                output = output.softmax(dim=1)

            predictions.extend(output.cpu().numpy())

    return predictions

test_predictions_folds = []
test_dataset = MelanomaDataset(
    df=test_df,
    meta_features=meta_variables,
    hdf=hdf_test_path,
    transform=get_rgb_transform(224, 224)[1],
    mode='test'
)
test_dataloader = get_dataloader(test_dataset, split='test')

n_meta_features = len(meta_variables) if meta_variables is not None else 0
use_meta = n_meta_features > 0

model_class = EffNetB3_Melanoma(out_dim=1, n_meta_features=n_meta_features)
model_class.to(CONFIG['device'])

for fold_idx in range(5):
    fold_checkpoint_path = f'/kaggle/input/dl-modeling/best_fold_{fold_idx + 1}.pth'
    model = load_model(model_class, fold_checkpoint_path, CONFIG['device'])
    test_predictions = inference(model, test_dataloader, CONFIG['device'], use_meta=use_meta)
    test_predictions_folds.append(test_predictions)

submission_predictions = np.mean(test_predictions_folds, axis=0)
```
